```python
import pyodbc 
from os import getcwd
import nltk

DRIVER_NAME = "Microsoft Access Driver (*.mdb, *.accdb)"
# Ruta completa del archivo.
#DB_PATH = getcwd() + "/Revisiones_literatura-DESKTOP-92GM44S.accdb"
DB_PATH="C:\\Users\\HP\\OneDrive\\Investigaciones\\Doctorado\\Revision sistemática\\Revision usando star office\\Revisiones_literatura-DESKTOP-92GM44S.accdb" 
# Establecer la conexión.
conn = pyodbc.connect("Driver={%s};DBQ=%s;" % (DRIVER_NAME, DB_PATH))


SQL_Query="SELECT id,Observaciones from Articulo"
cursor = conn.cursor()
cursor.execute(SQL_Query)
rows = cursor.fetchall()

if rows is not None:
    for row in rows:
        text = row.Observaciones
        if text is not None: 
            textTokenizer = nltk.Text(nltk.word_tokenize(text,'Spanish'))
            match = textTokenizer.concordance('Turismo')
            if match is not None:
                print(row.id, match)
       # print(row.id, row.Observaciones)

cursor.close()
conn.close()
```
